```python
import matplotlib.pyplot as plt
x=['Red', 'Purple', 'Orange', 'Green', 'Blue']
y=[1,1,1,2,4]
plt.plot (x, y, marker='o', linestyle='-', color='black')
plt.title ('Favorite Colors')
plt.xlabel('Colors')
plt.ylabel('Participants')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
Colors =['red','purple','orange','green','blue']
Participants = [1,1,1,2,4]
plt.bar(Colors, Participants, color='brown')
plt.title('Favorite Colors')
plt.xlabel('Colors')
plt.ylabel('Participants')
plt.show()
```
